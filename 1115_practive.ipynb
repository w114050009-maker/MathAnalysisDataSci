## 1.2 讀取 .csv

```python
#data = pd.read_csv('/content/drive/MyDrive/housing.csv')
data = pd.read_csv('/content/drive/MyDrive/housing.csv')
data
```

```python
data.info() #data.isnull().any()
```

```python
data.dropna(inplace=True) #drop 'not available' data from dataset
#data.dropna() #not drop
```

```python
data.info()
```

```python
x = np.array(data)[:, :8].astype(np.float32) #to column 7th with all rows
x
```

```python
y = np.array(data)[:, 8].astype(np.float32)
y
```

By LSM

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
lin_reg.fit(x, y)
```

```python
lin_reg.intercept_, lin_reg.coef_ #bias, weights
```

```python
lin_reg.predict(x[:3])
```

```python
from sklearn.preprocessing import StandardScaler

scal = StandardScaler()
x_scal = scal.fit_transform(x)

scal.mean_, scal.scale_
```

By LSM + Scaled

```python
lin_reg.fit(x_scal, y)

lin_reg.intercept_, lin_reg.coef_
```

```python
lin_reg.predict(x_scal[:3])
```

By SGD

```python
from sklearn.linear_model import SGDRegressor

sgd_reg = SGDRegressor(max_iter=1000, penalty=None, eta0=1e-3, tol=0.1)
sgd_reg.fit(x_scal, y.ravel())

sgd_reg.intercept_, sgd_reg.coef_
```

```python
sgd_reg.predict(scal.transform(x[:3]))
```
